### Formateando cadenas

```python
"Tiempo estimado: {:.2f}s".format(1.34844)
```

```python
"Valor para muestra {:05d}: |{:>13s}|".format(7, "failure")
```

### Accediendo a los caracteres

```python
cad = "Hola mundo"
cad[0]
```

```python
cad[2]
```

```python
cad[-2]
```

### Accediendo subcadenas

```python
cad = "Hola mundo"
cad[2:6]
```

```python
cad[:6]
```

```python
cad[2:]
```

```python
cad[-3:]
```
