### Devolviendo objetos

```python
def foo(a, b):
    return a + b
```

```python
r = foo(3, 4)
r
```

```python
def foo(a, b):
    mul = a * b
    if b == 0:
        return mul, None

    div = a / b
    return mul, div
```

```python
foo(3, -2)
```

```python
m, d = foo(3, -2)
```

```python
m
```

```python
d
```

```python
print("Resultados: multip={} divis={}".format(m, d))
```

```python
m, d = foo(3, 0)
print("Resultados: multip={} divis={}".format(m, d))
```

## Generadores

### Ejemplo inicial

```python
sum(range(10))
```

### Rango "casero", forma clásica

```python
def rango(limite):
    nros = []
    n = 0
    while n < limite:
        nros.append(n)
        n += 1
    return nros
```

```python
rango(10)
```

```python
sum(rango(10))
```

### Rango "casero", como generador

```python
def rango(limite):
    n = 0
    while n < limite:
        yield n
        n += 1
```

```python
g = rango(10)
g
```

```python
next(g)
```

```python
next(g)
```

```python
sum(rango(10))
```

### Generator comprehension

```python
(x ** 2 for x in range(3))
```

```python
sum([x ** 2 for x in range(10)])
```

```python
sum(x ** 2 for x in range(10))
```
