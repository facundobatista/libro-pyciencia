# Tipos de datos: Conjuntos

### Construyendo

```python
frutas = {'manzana', 'naranja', 'sandía', 'limón'}
frutas
```

```python
vitamina_c = {'limón', 'perejil', 'fresa', 'naranja'}
```

```python
vitamina_c
```

```python
frutas & vitamina_c
```

```python
frutas - vitamina_c
```

```python
frutas | vitamina_c
```

### Agregando objetos

```python
música = {'metal', 'tango', 'rock'}
música
```

```python
música.add('rock')
música
```

```python
música.add('jazz')
música
```

```python
música.update(['tango', 'cumbia'])
música
```

```python
'rock' in música
```

### Borrando objetos

```python
música.remove('jazz')
música
```

```python
música.remove('jazz')
```

```python
música.discard('metal')
música
```

```python
música.discard('metal')
música
```
