# Tipos de datos: Iteradores

### Iterando a mano

```python
números = [1, 2, 3]
type(números)
```

```python
iterador = iter(números)
type(iterador)
```

```python
next(iterador)
```

```python
next(iterador)
```

```python
next(iterador)
```

```python
next(iterador)
```

### Iterando con list()

```python
list([1, 2, 3])  # una lista
```

```python
list((1, 2, 3))  # una tupla
```

```python
list("abcd")  # una cadena
```

```python
list({1, 2, 3})  # un conjunto
```

```python
list({'a': 1, 'b': 2, 'c': 3})  # un diccionario
```

### Mezclado

```python
lista = [1, 2, 3, 4, 5, 6, 7]
list(lista)
```

```python
iterador = iter(lista)
iterador
```

```python
next(iterador)
```

```python
next(iterador)
```

```python
list(iterador)
```

### Iterando estructuras muy grandes

```python
nros = range(3)
list(nros)
```

```python
muchos_nros = range(1000000000000000)
iter_muchos_nros = iter(muchos_nros)
next(iter_muchos_nros)
```

```python
next(iter_muchos_nros)
```
