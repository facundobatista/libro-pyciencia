# Tipos de datos: Listas

### Definiendo, operando, accediendo

```python
lista = [1, "a", 5.6, "foo"]
lista
```

```python
len(lista)
```

```python
lista[1]
```

```python
lista[2:]
```

```python
lista * 2
```

```python
lista + [1, 2]
```

### Agregando y extendiendo

```python
lista = [1, 2, 3]
lista
```

```python
lista.append(4)
lista
```

```python
otra = [5, 6]
lista.extend(otra)
lista
```

### Reemplazando

```python
lista
```

```python
lista[1] = 99
lista
```

```python
lista[2:] = [0, 0]
lista
```

### Borrando

```python
lista
```

```python
lista.remove(99)
lista
```

```python
del lista[1]
lista
```

### Construyendo

```python
list("abcd")
```
