```python
def fibonacci(limit):  # función clásica
    numbers = [1]
    a, b = 0, 1
    while True:
        a, b = b, a + b 
        if b > limit:
            break
        numbers.append(b)
    return numbers

for n in fibonacci(10):
    print(n)
```

```python
def fibonacci(limit):  # generador
    yield 1
    a, b = 0, 1
    while True:
        a, b = b, a + b 
        if b > limit:
            break
        yield b

for n in fibonacci(10):
    print(n)
```

```python
def fibonacci():  # generador infinito
    yield 1
    a, b = 0, 1
    while True:
        a, b = b, a + b 
        yield b

for n in fibonacci():
    if n > 10:
        break
    print(n)
```

```python
def foo(simple):
    yield 1
    if simple:
        return
    yield 2
```

```python
list(foo(simple=False))
```

```python
def foo():
    a = yield 5
    print("en el generador:", a)
    yield a * 3
```

```python
g = foo()
val = next(g)
print("afuera:", val)
```

```python
val = g.send(3)
print("afuera:", val)
```
