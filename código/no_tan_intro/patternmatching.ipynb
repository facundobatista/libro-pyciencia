```python
def power_2(value, allowed_floats=False):
    match value:
        case int():
            print(2 ** value)
        case float() if allowed_floats:
            print(2 ** value)
        case _:
            print("ERROR")
      
```

```python
power_2(2)
```

```python
power_2(2.0)
```

```python
power_2(2.0, allowed_floats=True)
```

```python
power_2(None)
```
