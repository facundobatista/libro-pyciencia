```python
import sympy as sym

x, y = sym.symbols('x y')
sym.Eq(2 * x + 8, 16 * y)
```

```python
sym.solveset(sym.Eq(2 * x + 8, 16 * y), x)
```

```python
sym.solveset(sym.Eq(2 * x + 8 - 16 * y, 0), x)
```

```python
sym.solveset(2 * x + 8 - 16 * y, x)
```

```python
sym.solve(x ** 2 + 2 * x - 3)
```

```python
sym.solveset(x ** 2 + 2 * x - 3, x)
```

```python
a, b, c = sym.symbols('a b c')
sym.solveset(a *x ** 2 + b * x + c, x)
```

```python
sym.solve(sym.cos(x) - sym.sin(x), x)
```

```python
sym.solveset(sym.cos(x) - sym.sin(x), x)
```

```python
sym.solve(x ** 5 - x ** 2 + 1, x)
```

```python
sym.solveset(sym.tan(x) + x, x)
```

```python
sym.solve(sym.exp(x), x)
```

```python
sym.solveset(sym.exp(x), x)
```

```python
ecuación_1 = 2 * x - 5 * y + 4
ecuación_2 = -3 * x + 4 * y + 1
sym.solve([ecuación_1, ecuación_2], [x, y], dict=True)
```

```python
ecuación_1 = x ** 2 - 2 * y
ecuación_2 = y ** 2 - 2 * x 
soluciones = sym.solve([ecuación_1, ecuación_2], [x, y], dict=True)
soluciones
```

```python
[ecuación_1.subs(sol).simplify() == 0 and ecuación_2.subs(sol).simplify() == 0 
 for sol in soluciones]
```

```python
sym.solve(x ** 4 - 2 ** 4, x)
```

```python
sym.solveset(x ** 4 - 2 ** 4, x)
```

```python
x = sym.Symbol('x', real=True)
sym.solve(x ** 4 - 2 ** 4, x)
```

```python
polinomio = x ** 4 + 2 * x ** 3 - 25 * x ** 2 - 26 * x + 120
solución = sym.solve(polinomio, x)
solución
```

```python
sol_intervalo = [v for v in solución if (v.is_real and sym.Or(v < 0 , v > 3))]
sol_intervalo
```

```python
sym.solveset(x ** 4 - 2 ** 4, x, domain=sym.S.Reals)
```

```python
sym.solveset(polinomio, x, sym.Interval(0, 3))
```

```python
sym.solve(sym.cos(x) - x, x)
```

```python
sym.nsolve(sym.cos(x) - x, x, 1)
```
