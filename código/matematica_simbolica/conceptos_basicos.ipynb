```python
(x + 2) ** 2 == x ** 2 + 4 * x + 4
```

```python
izquierda = (x + 2) ** 2
derecha = x ** 2 + 4 * x + 4
sym.simplify(izquierda - derecha)
```

```python
derecha_2 = x ** 2 + 2 ** 2
sym.simplify(izquierda - derecha_2)
```

```python
print(izquierda.equals(derecha))
print(izquierda.equals(derecha_2))
```
