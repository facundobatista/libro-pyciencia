# Controles de flujo: If, elif, else

### Ejemplo inicial

```python
from datetime import date

hoy = date.today()
```

```python
if hoy >= date(2030, 1, 1): 
    print("El futuro ya llegó") 
elif date(2020, 1, 1) <= hoy < date(2030, 1, 1): 
    print('La famosa "década del 20"') 
else: 
    print("El lejano pasado") 
```
