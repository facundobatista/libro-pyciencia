# Controles de flujo: While

### Ejemplo inicial

```python
a = 0
while a < 3:
    a += 1
    print(a)
```

### El break

```python
a = 0
while a < 5:
    a += 1
    if a == 3:
        break
    print(a)
```

### El continue

```python
a = 0
while a < 5:
    a += 1
    if a == 3:
        continue
    print(a)
```

### El else

```python
a = 0
while a < 2:
    a += 1
    if a == 3:
        break  # realmente no va a llegar acá
    print(a)
else:
    print("en el else")
print("afuera")
```

```python
a = 0
while a < 4:  # ahora sí vamos a tocar el break
    a += 1
    if a == 3:
        break
    print(a)
else:
    print("en el else")
print("afuera")
```
