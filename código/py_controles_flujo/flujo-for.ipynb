# Controles de flujo: for

### Primer ejemplo

```python
nros = [1, 2, 3]
for n in nros:
    print(n)
```

### Asignación múltiple

```python
cosas = [('a', 1), ('b', 2), ('c', 3)]
for letra, nro in cosas:
    print(letra, nro ** 2)
```

### El range

```python
list(range(5))
```

```python
list(range(3, 8))
```

```python
list(range(3, 8, 2))
```

```python
list(range(3, 12, 2))
```

### Iterando el range

```python
for n in range(2, 5):
    print("El cuadrado de {} es {}".format(n, n ** 2))
```

### Break, continue y else

```python
números = [1, 2, 3, 4]
for n in números:
    if n == 3:
        break
    print(n)
```

```python
for n in números:
    if n == 3:
        continue
    print(n)
```

```python
for n in números:
    print(n)
else:
    print("se nos terminó el iterable")
```

```python
for n in números:
    if n == 2:
        break
    print(n)
else:
    print("se nos terminó el iterable")
```

### Comprensión de listas

```python
números = [1, 3, -2, 2, 0, 5]
cuadrados = []
for n in números:
    cuadrados.append(n ** 2)

cuadrados
```

```python
números = [1, 3, -2, 2, 0, 5]
cuadrados = [n ** 2 for n in números]
cuadrados
```

```python
import math
números = [1, 3, -2, 2, 0, 5]
logs1 = []
for n in números:
    if n > 0:
        logs1.append(math.log(n))

logs1
```

```python
logs2 = [math.log(n) for n in números if n > 0]
logs2
```

### Otras comprensiones

```python
números = [1, 3, -2, 2, 0, 5]
[n ** 2 for n in números]
```

```python
{n ** 2 for n in números}
```

```python
{n: n ** 2 for n in números}
```
