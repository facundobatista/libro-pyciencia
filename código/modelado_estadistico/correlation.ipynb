```python
import pandas as pd

columns = ['Sex', 'Age', 'Height', 'Weight', 'Year', 'Sport', 'NOC']
olymp = pd.read_csv('athlete_events.csv', usecols=columns).dropna()
olymp.head()
```

```python
olymp.info()
```

```python
import matplotlib.pyplot as plt

# Creamos los objetos figura y ejes
fig, ax = plt.subplots(1, 2,figsize=(10,5))

# Creamos los scatterplots
ax[0].plot(
    olymp['Height'], olymp['Weight'], marker='o', linestyle='', 
    alpha=0.1, markersize=2)
ax[1].plot(
    olymp['Age'], olymp['Weight'], marker='o', linestyle='', 
    alpha=0.1, markersize=2)

# Títulos y ejes
ax[0].set(
    title='Altura vs. Peso de atletas olímpicos',
    xlabel='Altura (cm)', ylabel='Peso (kg)', rasterized=True)
ax[1].set(
    title='Edad vs. Peso de atletas olímpicos',
    xlabel='Edad (años)', ylabel='Peso (kg)', rasterized=True)

plt.show()
```

```python
olymp[['Weight', 'Height', 'Age']].corr()
```

```python
from scipy.stats import rankdata

oly_10 = olymp.loc[:10, ['Weight', 'Height']]
oly_10['r_Weight'] = rankdata(oly_10['Weight'])
oly_10['r_Height'] = rankdata(oly_10['Height'])
oly_10['d'] = oly_10['r_Weight'] - oly_10['r_Height']
oly_10
```

```python
olymp[['Weight', 'Height', 'Age']].corr(method='spearman')
```
